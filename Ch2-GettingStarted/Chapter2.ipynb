```
"""Counting and Aggregating M&M"""
mnm_file="/databricks-datasets/learning-spark-v2/mnm_dataset.csv"
ip = spark.read.format('csv').option('header','true').option('inferSchema','true').load(mnm_file)
```

```
ip.show()
```

```
aggMNM=ip.select(ip['State'],ip['Color'],'Count').groupBy(ip['State'],ip['Color']).sum('Count')
aggMNM.show()
```

```
print('Count = %d' %aggMNM.count())
```

```
ca_count_mnm_df = aggMNM.select(aggMNM['State'],aggMNM['Color'],aggMNM['sum(Count)']).where(aggMNM['State']=='CA')
ca_count_mnm_df.show()
```
